# Node Capacity Check

## CCC Node

Refer to EC 2 Section 6.5.4 Fig. 6.26

### Design Parameters

```python
%%render params 2
k_1 = 1
b = 250*mm
a_1_1 = (1000*mm).prefix("m")
a_2_1 = (400*mm).prefix("m")
a_3_1 = (400*mm).prefix("m")
```

```python
%%render 2
A_1_1 = b*a_1_1
A_2_1 = b*a_2_1
A_3_1 = b*a_3_1
```

, where

b : Breadth of considered concrete section for node check

a<sub>i,j</sub> : Width of considered concrete section for node check

### Design Load (ULS)

```python
%%render params 2
F_Ed1_1 = (1000*kN).prefix("k")
F_Ed2_1 = (800*kN).prefix("k")
F_Ed3_1 = (600*kN).prefix("k")
```

```python
%%render 2
sigma_Ed1_1 = (F_Ed1_1/A_1_1).prefix("M")
sigma_Ed2_1 = (F_Ed2_1/A_2_1).prefix("M")
sigma_Ed3_1 = (F_Ed3_1/A_3_1).prefix("M")
```

### Node Capacity Check

Refer to EC 2 Section 6.5.4 Eq.(6.60),

```python
%%render 2
sigma_Rd_max = k_1*v*f_cd
```

### Demand-Capacity Ratio

```python
%%render 2
DCR_1_1 = sigma_Rd_max/sigma_Ed1_1
DCR_2_1 = sigma_Rd_max/sigma_Ed2_1
DCR_3_1 = sigma_Rd_max/sigma_Ed3_1
```
